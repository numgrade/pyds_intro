# Dataviz : introduction

La représentation graphique est une partie essentielle de l'analyse de données. Cela permet de comprendre ses données et de partager avec les autres ses analyses.

Aujourd'hui, le nombre d'outils pour visualiser ses données est très (trop?) impressionnant.

En voici une liste non-exhaustive :

[matplotlib](https://matplotlib.org/index.html) : La 1ère librairie qui a su s'imposer dans le monde Python. Son développement a commencé en 2002, elle était alors conçue comme une alternative à Matlab et elle est rapidement devenue la librairie la plus populaire dans le domaine de la représentation graphique en Python. Elle permet de produire des figures de qualité et même si sa syntaxe est parfois considérée comme bas-niveau ou compliquée, c'est probablement toujours la librairie la plus utilisée dans son domaine. Elle conserve une large communauté et son développement continue (version 3.1.3 en février 2020).

![image alt >](Images/numgrade.png)
Au cours de ces dernières années, plusieurs librairies ont été développées à partir de matplotlib et proposent une interface de plus haut niveau par rapport à matplotlib :

- [pandas plot](https://pandas.pydata.org/pandas-docs/stable/visualization.html) : pandas plot offre un moyen facile et rapide de représenter graphiquement les données issues d'un *DataFrame* pandas.

- [seaborn](http://seaborn.pydata.org/) : cette librairie est particulièrement adaptée pour la représentation de données statistiques avec tout un ensemble soigné de styles de tracé.

Nous pourrions également citer [ggpy](https://github.com/yhat/ggpy) et [plotnine](http://plotnine.readthedocs.io/en/stable/) : ces librairies ayant pour but de se rapprocher de la syntaxe et du style de la plus connue des librairies graphiques du langage R, à savoir ggplot2.

Les outils de représentation basés sur le Javascript (visualiser et interagir avec un graphe dans le navigateur ou les notebooks par exemple) :

- [bokeh](http://bokeh.pydata.org/en/latest/) : c'est une librairie de visualisation interactive qui cible les navigateurs web. Elle est très efficace et le rendu est remarquable mais elle ne comporte pas des fonctionnalités basiques telles que la génération de fichiers PDF.

- [plotly](https://plot.ly/python/) : plotly supporte plusieurs langages de programmation et le rendu 3D est agréable et dans le notebook, nous aurons une interactivité native. Cette librairie offre la possibilité de publier les graphiques en ligne dans des dashboards (option payante) ou d'exporter une version HTML/javascript sur le serveur de son choix.

![image alt >](Images/numgrade.png)
Une "nouvelle" alternative :

[Altair](https://github.com/ellisonbg/altair) : basé sur [Vega](https://vega.github.io/vega/) (une grammaire pour la visualisation), vous pouvez tracer un graphique à partir d'un *dataFrame* pandas. Le concept est différent de ce que l'on trouve pour les librairies de visualisation habituelles : vous ne vous souciez pas de la façon de tracer vos données, vous vous souciez uniquement de ce que vous souhaitez tracer. L'API Altair ne contient d'ailleurs pas de code pour le rendu du tracé mais produit une structure de données JSON (JavaScript Object Notation) qui pourra alors être tracée par tout logiciel pouvant interpréter ces données (le notebook sera capable d'interpréter ces données JSON). Un des avantages de ce concept est que vous pouvez facilement partager les données.

## Interface MATLAB-style Vs interface orientée objet:

Développée à l'origine comme une alternative à MATLAB, une des idées de départ était de retrouver une interface similaire à celle sous MATLAB. Mais lorsque les figures deviennent plus complexes, il est plus commode d'utiliser l'interface orientée objet qui permet également un contrôle plus important.

Comparons les 2 styles :

```python
# Old MATLAB style
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

![image alt >](Images/numgrade.png)

```python
# Object-oriented style
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

C'est l'interface orientée objet qui sera utilisée par la suite.

![image alt >](Images/numgrade.png)

## Comprendre les objets Figure et Axes de matplotlib

Dans l'exemple précédent, nous avons créé 2 variables :

- fig qui correspond à un objet [Figure ](https://matplotlib.org/stable/api/figure_api.html)

- et ax qui correspond à un tableau NumPy qui conient des objets [Axes](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.html)

On trouve une hiérarchie des objets dans matplotlib :

- l'objet **Figure** est le conteneur le plus externe d'un graphique matplotlib,

- le ou les objets **Axes** vont se trouver à l'intérieur de l'objet Figure (le nom Axes peut être un peu perturbant, mais un objet Axes correspond finalement à un graphique individuel),

- à l'intérieur de l'objet **Axes**, nous retrouverons une multitude de plus petits objets : **Axis**, **Title**, **Legends**, **Labels**, **Ticks**... Presque tous les éléments d'un graphique sont finalement représentés par des objets. Nous découvrirons plusieurs de ces objets dans la suite de la présentation.

Notez que Axes et Axis sont des objets différents.

## Titre, axe (étiquettes et graduations) et légende

Title, labels, ticks and legend

```python
fig, ax = plt.subplots(figsize = (6, 4))

# define a font size
font_size = 20

# Title 
ax.set_title("Sine", size=font_size)  # option size for the font size

# Labels
ax.set_xlabel("x", size=font_size)
ax.set_ylabel("y", size=font_size)

# Ticks - change the font size
ax.tick_params(labelsize=font_size)

# Axes limit
ax.axis([0, 3*np.pi, -1.5, 1.5])   # [xmin, xmax, ymin, ymax]

# plot data
x = np.linspace(0, 4* np.pi, 100)
ax.plot(x, np.sin(x), 'v', label='sine')
ax.plot(x, np.cos(x), 'rs', label='cosine')

# Legend
ax.legend(loc='best', prop={'size': font_size-4});
# or ax.legend(loc='best', fontsize=font_size-4);    # fontsize can also be used
# add a legend from the label options in ax.plot. 
# NOTE: ax.legend() should be placed after ax.plot 
# because the label has to be defined first. 
# The loc option means location. With 'best',
# matplolib will choose the best location.
```

## La méthode *set( )*

Il est possible de regrouper toutes les propriétés "set_title, set_xlabel, set_ylabel, set_xlim, set_ylim" dans la méthode *set()*:

```python
plt.style.use('seaborn-v0_8-whitegrid')   # add a grid 
fig, ax = plt.subplots(figsize = (6, 4))

# The set method is convenient but you lose control over font size
ax.set(xlim=(0, 3*np.pi), ylim=(-1.5, 1.5),
       xlabel='x', ylabel='y',
       title='sine - cosine')

# plot data
x = np.linspace(0, 4* np.pi, 100)
# Note the marker properties you can specify for example
ax.plot(x, np.sin(x), 'b-.', label = 'sine',
        markersize=10, markerfacecolor='green', 
        markeredgecolor='gray', markeredgewidth=2)
ax.plot(x, np.cos(x), 'r.', label= 'cosine')

ax.legend(loc='best');
```

Mais si avez besoin de contrôler les propriétés de chaque point, il sera indispensable d'utiliser *scatter( )*. Représentons avec un nuage de points le jeu de données sur l'iris que nous avions abordé dans la partie pandas :

```python
import pandas as pd
iris = pd.read_csv('data/iris.csv')
```

Nous avons 3 espèces différentes et 4 mesures pour chaque espèce.

```python
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(figsize = (7, 5))

# the c (color) option in plt.scatter() accept numbers with cmap, 
# so we transform the species names into numbers:
for i, specie in enumerate(iris['species'].unique()):
    iris.loc[iris['species']==specie, 'species'] = i

ax.scatter(iris['sepal_length'], iris['sepal_width'], alpha=0.5, # alpha: transparency
           s = 100*iris['petal_width'],  # s = size in pixel
           c = iris.loc[:, 'species'],            # c = colors
           cmap='jet',        # colormaps: https://matplotlib.org/users/colormaps.html
           )

ax.set_xlabel(iris.columns[0])
ax.set_ylabel(iris.columns[1]);
```

Avec ce graphique en nuage de points, nous sommes capable d'explorer les données dans 4 dimensions! Les coordonnées correspondent à la longueur et la largeur des sépales, la taille des points est reliée à la largeur des pétales et les couleurs correspondent aux 3 différentes espèces.

## Les barres d'erreur (error bar)

```python
plt.style.use('seaborn-v0_8-whitegrid')   # add a grid 
fig,ax = plt.subplots(figsize = (7, 5))

# plot data
x = np.linspace(0, 4* np.pi, 50)
y = np.sin(x)
dy = 0.1 + 0.2*np.sqrt(x)

ax.errorbar(x, y, yerr=dy);
```

![image alt >](Images/numgrade.png)
Vous pouvez bien sûr modifier le style des barres d'erreur :

```python
plt.style.use('seaborn-v0_8-whitegrid')   # add a grid 
fig,ax = plt.subplots(figsize = (7, 5))

# plot data
x = np.linspace(0, 4* np.pi, 50)
y = np.sin(x)
dy = 0.1 + 0.2*np.sqrt(x)

ax.errorbar(
    x, 
    y, 
    yerr=dy, 
    fmt='o-',   # style for the x, y plot
    color = 'black',
    ecolor = 'lightgray'  # color for the error bar
);
```

## Remplir une aire entre 2 lignes

Si on poursuit sur notre exemple précédent, nous pourrions avoir envie d'afficher une barre d'erreur continue (continuous error bars).

C'est possible avec la méthode *fill_between( )* :

```python
plt.style.use('seaborn-v0_8-whitegrid')   # add a grid 
fig,ax = plt.subplots(figsize = (7, 5))

# plot data
x = np.linspace(0, 4* np.pi, 50)
y = np.sin(x)
dy = 0.1 + 0.2*np.sqrt(x)

ax.plot(x, y, '-o')
ax.fill_between(x, y-dy, y+dy, alpha=0.2);
```

## Interagir avec les graphiques dans le Jupyter notebook

Dans les notebooks Jupyter, les [Jupyter widgets](http://jupyter.org/widgets.html) permettent de modifier les données d'entrée interactivement et de visualiser le résultat.

Pour interagir avec matplotlib, il est possible d'utiliser le widget [*interact*](http://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html) ([http://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html](http://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html).

Vous devez préalablement [installer](https://ipywidgets.readthedocs.io/en/stable/user_install.html) ce widget par exemple via pip :

  $ pip install ipywidgets

ou via conda :

   $ conda install -c conda-forge ipywidgets
 
Pour les anciennes versions de Jupyter Notebook ou JupyterLab, il fallait autoriser l'utilisation de l'extension via cette commande :

   $ jupyter nbextension enable --py widgetsnbextension
   
Cette dernière commande n'est normalement plus nécessaire.
    
Puis il faut importer le widget:

```python
import numpy as np
```

```python
%matplotlib inline
from ipywidgets import interactive, interact, Layout
import ipywidgets as widgets
```

Notez que vous devez utilisez "%matplotlib inline" ("%matplotlib notebook" ne fonctionne pas avec interact).

Une autre possibilité consiste à utiliser %matplotlib ipympl (ou %matplotlib widget - qui est équivalent) mais il faut alors avoir installé le package [ipympl](https://matplotlib.org/ipympl/#basic-example) (ce qui peut être intéressant avec ce package, c'est que vous pourrez également l'utiliser avec VS Code notebooks).

![image alt >](Images/numgrade.png)
Voyons un exemple avec des fonctions sinus et cosinus. Le widget va automatiquement créer une interface utilisateur pour explorer les données interactivement.

plt.style.use('seaborn-whitegrid')

```python
# To use the interactive function or the interact decorator, we have to define a function we want to explore
def functions(a, b, color_sin, color_cos, linewidth):
    fig, ax = plt.subplots(figsize=(10,7))
    
    font_size = 14
    # plot data
    x = np.linspace(0, 2*np.pi, 100)
    ax.plot(
        x, 
        a * np.sin(x + b), 
        color=color_sin, 
        linewidth=linewidth, 
        label='sine',
    )
    ax.plot(
        x, 
        a * np.cos(x + b), 
        color=color_cos, 
        linewidth=linewidth, 
        label='cosine',
    )
    
    # y limit
    ax.set_ylim(-2, 2)
    
    # Legend
    ax.legend(loc='best', fontsize= font_size)
    plt.show()

    
# interact autogenerates UI controls for function arguments, 
# and then calls the function with those arguments when you 
# manipulate the controls interactively
# You pass the function as the 1st argument, then the keyword arguments
interactive_plot = interactive(
    functions, 
    a=(0, 2.0), 
    b=(-np.pi, np.pi), 
    color_sin=['blue', 'green'],
    color_cos=['red', 'orange'],
    linewidth=widgets.IntSlider(min=1, max=4, step=1, value=2)
)

# You can specify the height of the plot with these 2 lines:
output = interactive_plot.children[-1]
output.layout.height = '410px'
# Show the plot:
interactive_plot
```

![image alt >](Images/numgrade.png)
Plus simplement, vous pourrez utiliser le décorateur @interact :

```python
@interact(
    a=(0, 2.0), 
    b=(-np.pi, np.pi), 
    color_sin=['blue', 'green'],
    color_cos=['red', 'orange'],
    linewidth=widgets.IntSlider(min=1, max=4, step=1, value=2),
)
def functions(a, b, color_sin, color_cos, linewidth):
    fig, ax = plt.subplots(figsize=(10,7))
    
    font_size = 14
    # plot data
    x = np.linspace(0, 2*np.pi, 100)
    ax.plot(
        x, 
        a * np.sin(x + b), 
        color=color_sin, 
        linewidth=linewidth, 
        label='sine',
    )
    ax.plot(
        x, 
        a * np.cos(x + b), 
        color=color_cos, 
        linewidth=linewidth, 
        label='cosine',
    )
    
    # y limit
    ax.set_ylim(-2, 2)
    
    # tick size
    ax.tick_params(labelsize=font_size)
    ax.tick_params(labelsize=font_size)
    
    # Legend
    ax.legend(loc='best', prop={'size': font_size})
    
    plt.show()
```

## Modifier les graduations et les axes
ticks, tick labels et spines

Partons du rendu standard de matplolib :

```python
plt.style.use('seaborn-v0_8-white')   # add a grid 

fig, ax = plt.subplots(figsize = (7, 5))
# data
x = np.linspace(-np.pi, np.pi, 100)
ax.plot(x, np.sin(x), 'b-', linewidth=3, label='sine')
ax.plot(x, np.cos(x), 'r-', linewidth=3, label='cosine')
# modify the limits so there is some space around the plots
ax.set_xlim(1.1 * x.min(), 1.1 * x.max())
ax.set_ylim(1.1 * np.min(np.sin(x)), 1.1 * np.max(np.sin(x)))
ax.legend(loc='best', prop={'size': 14}); 
```

Et maintenant, changeons les étiquettes et graduations (ticks, tick labels), modifions la position des axes (spines).

![image alt >](Images/numgrade.png)
##### Graduation et étiquettes

Nous allons utiliser les méthodes *set_xticks(), set_yticks(), set_xticklabels() et set_yticklabels()* :

```python
# Tick labels
# We first list the ticks and then precise the corresponding tick labels
font_size = 16
ax.tick_params(labelsize = font_size)
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels(['$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$'])

ax.set_yticks([-1, +1])
ax.set_yticklabels(['$-1$', '$+1$'])

fig
```

##### Positions des axes avec spines

```python
# Removing the top and right spines by setting the color to None
ax.spines['right'].set_color('None')
ax.spines['top'].set_color('None')
# Moving the bottom and left spines to coordinate 0
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# I want to see the major ticks - 
# options direction and length (and not for top and right ticks 
# because we removed top and right spines)
ax.tick_params(labelsize = font_size, 
               direction='inout', length=5, top=False, right=False)

fig
```


Nous pouvons remarquer que les "tick labels" ne sont plus très visibles. Essayons d'améliorer cette situation !

![image alt >](Images/numgrade.png)
Nous allons devoir modifier l'ordre de superposition des éléments avec "zorder".
Par défaut l'ordre de superposition des éléments est le suivant :


   |   element    |  Z-order |  
   |-----|----------|
   | Patch |  1       |
   | Line  |  2       |
   | Text  |  3       |


Le texte est au-dessus des lignes qui est au-dessus du reste.

Nous voulons au contraire que les courbes ne masquent pas les axes et les "tick labels". 
Nous allons donc modifier les "zorder" en profitant du fait que nous pouvons utiliser des floats. Nous prenons un zorder=0.1 pour les courbes (sinus et cosinus) et un zorder=0.2 pour les spines. Les tick labels apparaîtront alors au-dessus des courbes et des axes (spines).

Nous ajoutons également une boîte semi-transparente autour des tick labels pour bien voir les graduations et les courbes. 

```python
fig,ax = plt.subplots(figsize = (7, 5))
# data
x = np.linspace(-np.pi, np.pi, 100)
ax.plot(x, np.sin(x), 'b-', linewidth=3, label='sine', zorder=0.1)  ###
ax.plot(x, np.cos(x), 'r-', linewidth=3, label='cosine', zorder=0.1)  ###
# modify the limits so there is some space around the plots
ax.set_xlim(1.1 * x.min(), 1.1 * x.max())
ax.set_ylim(1.1 * np.min(np.sin(x)), 1.1 * np.max(np.sin(x)))
ax.legend(loc='best', prop={'size': 14})
 
# Spines
# Removing the top and right spine by setting the color to none
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# Moving the bottom and left spines to coordinate 0
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_zorder(0.2)                                    ###

# Tick labels
# We first list the ticks and then precise the corresponding tick labels
font_size = 16
ax.tick_params(labelsize = font_size, 
               direction='inout', length=5, top=False, right=False)
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels(['$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$'])
ax.set_yticks([-1, +1])
ax.set_yticklabels(['$-1$', '$+1$'])

###
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.6))
    
fig.savefig('sin_cos.pdf')
```
